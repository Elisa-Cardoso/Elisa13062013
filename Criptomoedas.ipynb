```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')  # Define o estilo do fundo como preto


# Resetando o √≠ndice para tornar 'Date' uma coluna regular
bitcoin.reset_index(inplace=True)
ethereum.reset_index(inplace=True)
litecoin.reset_index(inplace=True)
xrp.reset_index(inplace=True)

# Selecionando as colunas relevantes
df1_selected = bitcoin[['Date', 'Close_bitcoin']]
df2_selected = ethereum[['Date', 'Close_ethereum']]
df3_selected = litecoin[['Date', 'Close_litecoin']]
df4_selected = xrp[['Date', 'Close_xrp']]

# Realizando os merges
merged_df = pd.merge(df1_selected, df2_selected, on='Date', how='inner')
merged_df = pd.merge(merged_df, df3_selected, on='Date', how='inner')
merged_df = pd.merge(merged_df, df4_selected, on='Date', how='inner')

# Renomeando as colunas para facilitar a plotagem
merged_df.columns = ['Date', 'Bitcoin', 'Ethereum', 'Litecoin', 'XRP']

# Convertendo a coluna 'Date' para o formato datetime
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Configurando 'Date' como √≠ndice
merged_df.set_index('Date', inplace=False)

# Plotando a s√©rie temporal
plt.figure(figsize=(10, 6))
plt.plot(merged_df.index, merged_df['Bitcoin'], label='Bitcoin')
plt.plot(merged_df.index, merged_df['Ethereum'], label='Ethereum')
plt.plot(merged_df.index, merged_df['Litecoin'], label='Litecoin')
plt.plot(merged_df.index, merged_df['XRP'], label='XRP')

plt.title('S√©rie Temporal de Fechamentos')
plt.xlabel('Data')
plt.ylabel('Pre√ßo de Fechamento')
plt.legend()
plt.show()
```
